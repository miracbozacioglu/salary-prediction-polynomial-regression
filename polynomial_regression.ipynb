```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
dataset=pd.read_csv('Position_Salaries.csv')
X=dataset.iloc[:,1:-1].values
y=dataset.iloc[:,-1].values
```

```python
dataset.head()
```

```python
dataset.describe()
```

```python
plt.scatter(X,y,color='blue')
plt.title('Polynomial Regression')
plt.xlabel('Position Level')
plt.ylabel('Salary')
plt.show()
```

```python
from sklearn.linear_model import LinearRegression
lin_reg=LinearRegression()
lin_reg.fit(X,y)
```

```python
from sklearn.preprocessing import PolynomialFeatures
poly_reg=PolynomialFeatures(degree=4)
X_poly=poly_reg.fit_transform(X)
lin_reg_2=LinearRegression()
lin_reg_2.fit(X_poly,y)
```

```python
plt.scatter(X,y,color='blue')
plt.plot(X,lin_reg.predict(X),color='red')
plt.title('Linear Regression')
plt.xlabel('Position Level')
plt.ylabel('Salary')
plt.show()
```

```python
plt.scatter(X,y,color='blue')
plt.plot(X,lin_reg_2.predict(X_poly),color='red')
plt.title('Polynomial Regression')
plt.xlabel('Position Level')
plt.ylabel('Salary')
plt.show()
```

```python
from sklearn.metrics import mean_squared_error, r2_score
y_pred_lin=lin_reg.predict(X)
y_pred_poly=lin_reg_2.predict(X_poly)
mse_lin=mean_squared_error(y,y_pred_lin)
r2_lin=r2_score(y,y_pred_lin)
mse_poly=mean_squared_error(y,y_pred_poly)
r2_poly=r2_score(y,y_pred_poly)
```

```python
print("--- Model Evaluation Metrics ---")
print(f"Linear Regression - R-Squared Score: {r2_lin:.4f}")
print(f"Linear Regression - Mean Squared Error (MSE): {mse_lin:.2f}")

print(f"Polynomial Regression - R-Squared Score: {r2_poly:.4f}")
print(f"Polynomial Regression - Mean Squared Error (MSE): {mse_poly:.2f}")
```
